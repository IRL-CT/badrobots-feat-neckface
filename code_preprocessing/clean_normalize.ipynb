```python
#imports for standartizing dataset
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
#open dataset
path = '../data/training_data.csv'

data = pd.read_csv(path)
data
```

```python
#remove nans and infs
print(data.shape)
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna()
print(data.shape)
#save
data.to_csv('../data/training_data.csv', index=False)
```

```python
#normalize data
scaler = StandardScaler()

features = data.iloc[:, 5:]
print(features.head())
features = scaler.fit_transform(features)
features = pd.DataFrame(features)


data.iloc[:, 5:] = features
print(data.head())
print(data.shape)
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna()
print(data.shape)
 
#save normalized data
data.to_csv('../data/training_data_norm.csv', index=False)

```
